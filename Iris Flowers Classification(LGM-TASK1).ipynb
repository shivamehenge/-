```python
#seperat features and target
data = df.values
X = data[:,0:4]
Y = data[:,4]
```

# Train and Test Data

```python
#split the data to train and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2)
```

# Support Vector Machine

```python
#support vector machine algorithm
from sklearn.svm import SVC
svn = SVC()
svn.fit(X_train, Y_train)
```

```python
#predict from the dataset
predictions = svn.predict(X_test)
```

# accuracy of data

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, predictions)
```

# Classification Report

```python
#distibuted the classification report
from sklearn.metrics import classification_report
classification_report(Y_test, predictions)
```

# Test Model

```python
#Test the model
x_new = np.array([[2,3,5,5],[1,2,3.8,0],[5.3,2.5,4.6,1.9]])
#prediction of the species from the input vector
predictions = svn.predict(x_new)
print("Predictions of Species: {}".format(predictions))
```
